# Burger's equation

```python
import matplotlib.pyplot      as plt
import numpy                  as np
```

```python
# Space
L = 10.0     
nx = 101    
dx = L/(nx-1)   
x = np.linspace(0, L, nx)

# Define the initial condition
u_initial = 10*np.exp(-(x-5)**2)

# Time
cf = 0.5
t_final = 0.5
dt = cf * dx 
t_final_values = [t_final/2,t_final]

print('Number of iterations=',t_final/dt)

#GLOBAL VARIABLE
u = u_initial
```

```python
solutions_nc = []
solutions_c = []
```

```python
def evolve_conservative() :
    global u
    u_next = u - dt/(dx)*(0.5*u**2 - 0.5*np.roll(u, 1)**2)
    u = u_next
    return u

def evolve_noncons() :
    global u
    u_next =  u - u*dt/(dx)*(u - np.roll(u, 1)) 
    u = u_next
    return u
```

# Conservative

```python
u = u_initial
t = 0

while t <= t_final :
    
    if any(abs(t - val) < dt for val in t_final_values):
        # Append the current solution to the solutions list
        solutions_nc.append(u.copy())
        
    evolve_conservative()
    plt.close()
    t = t + dt

u_conservative = u
```

```python
plt.ylim(0,10)
plt.plot(x,u_conservative)
plt.show()
```

# Non conservative

```python
u = u_initial
t = 0

while t <= t_final :
    
    # Check if t is in t_final_values
    if any(abs(t - val) < dt for val in t_final_values):
        # Append the current solution to the solutions list
        solutions_c.append(u.copy())
        
    evolve_noncons()
    t = t + dt

u_non_conservative = u
```

```python
#plt.ylim(0,10)
plt.plot(x,u_non_conservative)
plt.show()
```

```python
plt.plot(x, u_initial, label='Initial')
plt.plot(x,u_non_conservative,label='Conservative')
plt.plot(x,u_conservative,label='Non Conservative')
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
plt.show()
```

# Plots

```python
# Plot the final solution
plt.title('Upwind flux-conservative Scheme Burger equation')
plt.plot(x, u_initial, label='Initial')

plt.plot(x, solutions_c[0], label='Time=' + str(t_final_values[0]))
plt.plot(x, solutions_c[1], label='Time=' + str(t_final_values[1]))

#plt.ylim(0, 1.1)
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
#plt.savefig('Burger FC.png')
plt.show()
```

```python
# Plot the final solution
plt.title('Upwind non flux-conservative Scheme Burger equation')
plt.plot(x, u_initial, label='Initial')

plt.plot(x, solutions_nc[0], label='Time=' + str(t_final_values[0]))
plt.plot(x, solutions_nc[1], label='Time=' + str(t_final_values[1]))

#plt.ylim(0, 1.1)
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
#plt.savefig('Burger FNC.png')
plt.show()
```
