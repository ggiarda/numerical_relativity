# Riemann Problem

```python
import matplotlib.pyplot as plt
import numpy as np
from kuibit.simdir import SimDir
from kuibit.grid_data import UniformGrid
```

## Exact Solution

```python
#initial data from the Exact Riemann solver of https://www.brunogiacomazzo.org/?page_id=395
exact = np.loadtxt("Sod_exact.dat")
```

```python
x_exact=exact[:,0]
rho_exact=exact[:,1]
press_exact=exact[:,2] #note: this is pgas+pmag
vx_exact=exact[:,3]
vy_exact=exact[:,4]
vz_exact=exact[:,5]
By_exact=exact[:,6]
Bz_exact=exact[:,7]
```

```python
plt.plot(x_exact, rho_exact, label="exact")
plt.xlim(-0.5, 0.5)
plt.legend()
plt.show()
```

## Read the Numerical Results

```python
gf= SimDir("./Sod_1d_400/").gf
print(gf)
```

```python
vars1d=gf.x
print(vars1d)
```

```python
rho=vars1d.fields.rho
```

## 2nd Resolution import

```python
gf2= SimDir("./Sod_1d_2/").gf
vars1d_2=gf2.x
print(vars1d_2)
rho2=vars1d_2.fields.rho
```

```python
plt.plot(rho2[400][0][0].coordinates()[0].data,rho2[400][0][0].data, label="GRHydro, dx = 0.00125")
plt.plot(x_exact, rho_exact, label="Exact")
plt.xlim(-0.5, 0.5)
plt.legend()
plt.show()
```

## 3rd Resolution import

```python
gf3= SimDir("./Sod_1d_3/").gf
vars1d_3=gf3.x
print(vars1d_3)
rho3=vars1d_3.fields.rho
```

```python
plt.plot(rho3[100][0][0].coordinates()[0].data,rho3[100][0][0].data, label="dx = 0.005")
plt.plot(x_exact, rho_exact, label="Exact")
plt.xlim(-0.5, 0.5)
plt.legend()
plt.show()
```

# Final Plot

```python
plt.plot(rho[200][0][0].coordinates()[0].data,rho[200][0][0].data, label="dx = 0.0025")
plt.plot(rho2[400][0][0].coordinates()[0].data,rho2[400][0][0].data, label="dx = 0.00125")
plt.plot(rho3[100][0][0].coordinates()[0].data,rho3[100][0][0].data, label="dx = 0.005")
plt.plot(x_exact, rho_exact, label="Exact")
plt.xlim(-0.5, 0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$\rho$')
plt.legend()
plt.savefig("Sod Tube.png")
plt.show()
```
